```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
X=df[['Area']]
y=df['Price']
model = LinearRegression()
model.fit(X,y)
predictions = model.predict(X)
predictions
```

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, predictions, color='red', linewidth=2, label='Linear Regression Fit')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression Fit on Data')
plt.legend()
plt.grid(True)
plt.show()
```
