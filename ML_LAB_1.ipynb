```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the Excel file (assuming it's uploaded to Colab)
df = pd.read_csv('Ml_LAB_AREA.csv')

# Display first 5 rows
print(df.head())
```

```python
# Set features and target
X = df[['Area (sq ft)']]
y = df['Price ']  # Or 'Price ($)' if that's your column name
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict prices
predicted_prices = model.predict(X)
```

```python
# Plot the data and the regression line
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='blue', label='Actual Prices')
plt.plot(X, predicted_prices, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression: House Price vs. Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ')  # Adjust label according to your currency
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Print model parameters
print(f"\nModel Intercept: {model.intercept_}")
print(f"Model Coefficient (Slope): {model.coef_[0]}")
```
